# Sección SQLite

```python
#Importamos SQlite
import sqlite3
```

```python
# Conectar a la base de datos
conn = sqlite3.connect('ecommerce.db')
cursor = conn.cursor()
#Esto crea un archivo.db (ecommerce) dentro de content.

# Guardar los df como tablas en la base de datos ecommerce.db
df_orders.to_sql('orders', conn, if_exists='replace', index=True)
df_customers.to_sql('customers', conn, if_exists='replace', index=True)
df_orderPayments.to_sql('order_payments', conn, if_exists='replace', index=True)
df_orderItems.to_sql('order_items', conn, if_exists='replace', index=True)
df_products.to_sql('products', conn, if_exists='replace', index=True)

```

## PRAGMA: Ver estructuras y tipo de datos de las tablas
Guardado en archivos .info

```python
#Ejecutar una consulta PRAGMA para obtener la estructura de las tablas y guardar la info a un archivo con extensión .info.


#Se muestra estructura y tipo de dato de tabla customers
prag_customers =  pd.read_sql_query('''PRAGMA table_info(customers);
''', conn)
#Se muestra estructura y tipo de dato de tabla orders
prag_orders =  pd.read_sql_query('''PRAGMA table_info(orders);
''', conn)
#Se muestra estructura y tipo de dato de tabla order_items
prag_orderItems =  pd.read_sql_query('''PRAGMA table_info(order_items);
''', conn)
#Se muestra estructura y tipo de dato de tabla order_payments
prag_orderPayments =  pd.read_sql_query('''PRAGMA table_info(order_payments);
''', conn)
#Se muestra estructura y tipo de dato de tabla products
prag_products =  pd.read_sql_query('''PRAGMA table_info(products);
''', conn)

#Se guardan en archivos .info
prag_customers.to_csv('customers_info.info', index=False)
prag_orders.to_csv('orders_info.info', index=False)
prag_orderItems.to_csv('orderItems_info.info', index=False)
prag_orderPayments.to_csv('orderPayments_info.info', index=False)
prag_products.to_csv('products_info.info', index=False)
```

## PRAGMA : Índices de las tablas
Guardado en archivos .info

```python
#Obtener los nombres de los índices de las tablas y guardar la info a un archivo con extensión .info.

#Se muestra el índice de tabla customers
index_customers =  pd.read_sql_query('''PRAGMA index_list(customers);
''', conn)
index_info_customers = pd.read_sql_query('''PRAGMA index_info(ix_customers_customer_id);
''', conn)

#Se guardan en un archivo .info
index_customers.to_csv('index_customers.info', index=False)
index_info_customers.to_csv('index_info_customers.info', index=False)

#Se muestra el índice de tabla order_items
index_orderItems =  pd.read_sql_query('''PRAGMA index_list(order_items);
''', conn)
index_info_orderitems = pd.read_sql_query('''PRAGMA index_info(ix_order_items_order_id);
''', conn)

#Se guardan en un archivo .info
index_orderItems.to_csv('index_orderItems.info', index=False)
index_info_orderitems.to_csv('index_info_orderitems.info', index=False)

#Se muestra el índice de tabla order_payments
index_orderPayments =  pd.read_sql_query('''PRAGMA index_list(order_payments);
''', conn)
index_info_orderPayments = pd.read_sql_query('''PRAGMA index_info(ix_order_payments_order_id);
''', conn)

#Se guardan en un archivo .info
index_orderPayments.to_csv('index_orderPayments.info', index=False)
index_info_orderPayments.to_csv('index_info_orderPayments.info', index=False)

#Se muestra el índice de tabla orders
index_orders =  pd.read_sql_query('''PRAGMA index_list(orders);
''', conn)
index_info_orders = pd.read_sql_query('''PRAGMA index_info(ix_orders_order_id);
''', conn)

#Se guardan en un archivo .info
index_orders.to_csv('index_orders.info', index=False)
index_info_orders.to_csv('index_info_orders.info', index=False)

#Se muestra el índice de tabla products
index_products =  pd.read_sql_query('''PRAGMA index_list(products);
''', conn)
index_info_products = pd.read_sql_query('''PRAGMA index_info(ix_products_product_id);
''', conn)

#Se guardan en un archivo .info
index_products.to_csv('index_products.info', index=False)
index_info_products.to_csv('index_info_products.info', index=False)

```

### Valor promedio de los pedidos por cliente

```python

#Calcula el valor promedio de los pedidos realizados por cada cliente.

query_2 = pd.read_sql_query('''SELECT customers.customer_id, avg( distinct order_items.price) AS avg_order_value
FROM customers
JOIN orders ON customers.customer_id = orders.customer_id
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY customers.customer_id
HAVING COUNT(orders.order_id) > 1
ORDER BY avg_order_value DESC
LIMIT 10;
''', conn)

avg_x_cliente = query_2


# Exporta el DataFrame a un archivo CSV
avg_x_cliente.to_csv('avg_x_cliente.csv', index=False)
```

### Clientes mas frecuentes y su promedio de frecuencia de compra

```python
 #Mostrar los clientes mas frecuentes y con su promedio de compra
clientes_frec = pd.read_sql_query('''SELECT
    c.customer_id,
    c.customer_unique_id,
    COUNT(o.customer_id) AS num_orders,
    round(AVG(oi.price + oi.freight_value), 2) AS avg_purchase_amount
FROM
    customers c
JOIN
    orders o ON c.customer_id = o.customer_id
JOIN
    order_items oi ON o.order_id = oi.order_id
GROUP BY
    c.customer_id, c.customer_unique_id
ORDER BY
    num_orders DESC
LIMIT 20;''', conn)

clientes_frec

# Exporta el DataFrame a un archivo CSV
clientes_frec.to_csv('clientes_frec.csv', index=False)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

# Gráfico de barras para el número de pedidos
sns.barplot(x='num_orders', y='customer_id', data=clientes_frec, color='blueviolet', label='Número de Pedidos')


plt.title('Clientes Más Frecuentes y su Promedio de Compra - Pedidos')
plt.xlabel('Pedidos')
plt.ylabel('Customer ID')
plt.legend()
plt.show()
```

```python
# Gráfico de barras para el promedio de compra
sns.barplot(x='avg_purchase_amount', y='customer_id', data=clientes_frec, color='darkblue', label='Promedio de Compra',
            order=clientes_frec.sort_values('avg_purchase_amount', ascending=False)['customer_id'])

plt.title('Clientes Más Frecuentes y su Promedio de Compra -Promedio de Compra')
plt.xlabel('Promedio (Valor)')
plt.ylabel('Customer ID')
plt.legend()
plt.show()
```

# Análisis de popularidad de productos por categoría
Se utilizan las tablas products y order_items

```python
#Agrupar los productos por categoría y calcula la cantidad total de unidades vendidas para cada categoría.
ord_x_cat =  pd.read_sql_query('''SELECT products.product_id, products.product_category_name,
count(order_items.product_id) as qty_orders, round(sum(order_items.price),2) as suma_vtas
FROM products
JOIN order_items ON products.product_id = order_items.product_id
GROUP BY products.product_category_name
HAVING qty_orders > 2000
ORDER BY qty_orders DESC
''', conn)
ord_x_cat

# Exporta el DataFrame a un archivo CSV
ord_x_cat.to_csv('ord_x_cat.csv', index=False)
```

### Categorías populares (medido en ventas)

```python

#Generar un informe que muestre las categorías de productos más populares en términos de ventas.

vtas_x_cat =  pd.read_sql_query('''SELECT products.product_id, products.product_category_name, count(order_items.product_id) as qty_orders, round(sum(order_items.price),2) as suma_vtas
FROM products
JOIN order_items ON products.product_id = order_items.product_id
GROUP BY products.product_category_name
HAVING suma_vtas > 400000
ORDER BY suma_vtas DESC
''', conn)
vtas_x_cat

# Exporta el DataFrame a un archivo CSV
vtas_x_cat.to_csv('vtas_x_cat.csv', index=False)
```

```python
#Calcula el tiempo promedio de entrega de los pedidos por cliente.
avg_entrega =  pd.read_sql_query('''
SELECT
    customer_id,
    ROUND(AVG(julianday(order_delivered_customer_date) - julianday(order_purchase_timestamp)), 2) as avg_order_days
FROM orders
WHERE order_delivered_customer_date IS NOT NULL
GROUP BY customer_id
ORDER BY avg_order_days desc
;
''', conn)

avg_entrega

# Exporta el DataFrame a un archivo CSV
avg_entrega.to_csv('avg_entrega.csv', index=False)
```

```python

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

# Gráfico de violín para el tiempo promedio de entrega
sns.violinplot(x=avg_entrega['avg_order_days'], color='skyblue')

plt.title('Distribución del Tiempo Promedio de Entrega por Cliente')
plt.xlabel('Días Promedio de Entrega')
plt.ylabel('Densidad de Datos')

plt.show()

```

```python
#Promedio de velocidad de entrega medido en días (total, no por cliente)
avg_entrega_dias =  pd.read_sql_query('''
SELECT
    CAST(AVG(julianday(order_delivered_customer_date) - julianday(order_purchase_timestamp)) AS INT) as avg_order_days
FROM orders''', conn)

avg_entrega_dias

# Exporta el DataFrame a un archivo CSV
avg_entrega_dias.to_csv('avg_entrega_dias.csv', index=False)
```

### Tiempos mas largos y mas cortos de entrega


```python
#Clientes con los tiempos de entrega más lentos

max_time = pd.read_sql_query('''
SELECT
      c.customer_id,
      round(max(julianday(o.order_delivered_customer_date) - julianday(o.order_purchase_timestamp)), 2) as max_order_days

FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
WHERE order_delivered_customer_date IS NOT NULL
GROUP BY c.customer_id
ORDER BY max_order_days desc
LIMIT 10;
''', conn
)
max_time

# Exporta el DataFrame a un archivo CSV
max_time.to_csv('max_time.csv', index=False)
```

```python
#Clientes con los tiempos de entrega más rápidos

min_time = pd.read_sql_query('''
SELECT
      c.customer_id,
      round(min(julianday(o.order_delivered_customer_date) - julianday(o.order_purchase_timestamp)), 2) * 60 as min_order_minutes

FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
WHERE order_delivered_customer_date IS NOT NULL
GROUP BY c.customer_id
ORDER BY min_order_minutes
LIMIT 10;
''', conn
)
min_time

# Exporta el DataFrame a un archivo CSV
min_time.to_csv('min_time.csv', index=False)
```

```python
# Datos de los tiempos máximos y mínimos
clientes_max = max_time['customer_id'].tolist()
tiempos_max = max_time['max_order_days'].tolist()

clientes_min = min_time['customer_id'].tolist()
tiempos_min = min_time['min_order_minutes'].tolist()

plt.figure(figsize=(12, 6))

# Gráfico de barras para los clientes con tiempos máximos
plt.bar(clientes_max, tiempos_max, color='m', label='Tiempo Máximo (días)')

# Gráfico de barras para los clientes con tiempos mínimos
plt.bar(clientes_min, tiempos_min, color='c', label='Tiempo Mínimo (minutos)')

plt.title('Clientes con Tiempos de Entrega (Max y Min)')
plt.xlabel('Cliente')
plt.ylabel('Tiempo de Entrega')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas para mejor legibilidad
plt.legend()
plt.tight_layout()

plt.show()












```

### Agrupar los pagos por método. Calcular total de pago por método

```python
#Agrupar los pagos por método de pago y calcula el total de pagos realizados con cada método.
total_x_metodo = pd.read_sql_query('''
SELECT
       payment_type, round(sum(payment_value), 2) as total
FROM order_payments
WHERE payment_type <> 'not_defined'
GROUP BY payment_type
ORDER BY total DESC
''', conn
)
total_x_metodo

# Exporta el DataFrame a un archivo CSV
total_x_metodo.to_csv('total_x_metodo.csv', index=False)
```

```python
# Datos para el gráfico de pastel
metodos_pago = total_x_metodo['payment_type'].tolist()
totales = total_x_metodo['total'].tolist()

# Paleta de colores personalizada
colores = sns.color_palette("Set3", len(metodos_pago))

plt.figure(figsize=(5, 5))

# Crear un gráfico de pastel
sns.set(style="whitegrid")
plt.pie(totales, labels=None, colors=colores, startangle=140, autopct='')

plt.title('Distribución de Pagos por Método de Pago')

plt.axis('equal')
plt.legend(metodos_pago, title="Método de Pago", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))  # Agregar leyenda

plt.show()
```

Informe métodos de pago más utilizados


```python
#Métodos de pago más utilizados por los clientes.

metodo_popular = pd.read_sql_query('''
SELECT
       payment_type, count(payment_type) as method
FROM order_payments
WHERE payment_type <> 'not_defined'
GROUP BY payment_type
ORDER BY method DESC
''', conn
)
metodo_popular

# Exporta el DataFrame a un archivo CSV
metodo_popular.to_csv('metodo_popular.csv', index=False)
```

```python
plt.figure(figsize=(8, 6))

# Crear un gráfico de barras en Seaborn
sns.set(style="whitegrid")
sns.barplot(x='method', y='payment_type', data=metodo_popular, palette="viridis")

plt.title('Método de Pago Más Popular')
plt.xlabel('Cantidad de Pagos')
plt.ylabel('Método de Pago')

plt.show()
```
